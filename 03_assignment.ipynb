김철수씨는 최근에 어느 광산에 취직했습니다.

그 광산엔 금, 은, 다이아몬드가 있다고 합니다.

김철수씨의 목표는 금, 은, 다이아몬드를 각각 1개씩 채굴하는 것이고

김철수씨를 광부 클래스의 객체로 생성하고 위의 문제를 해결하세요.


```python

# 실행예시

# 목표=[금,은,다이아몬드] 순서로 대입합니다. 즉, 아래 목표는 금, 은, 다이아를 각각 1개씩으로 하는 것을 목표로 한 것입니다.

김철수 = 광부(이름='김철수', 목표=[1,1,1])

김철수.채굴()

    무슨 광물을 채굴할까? 다이아몬드
    뚝딱뚝뚝....(3초)
    다이아몬드을(를) 채굴했습니다.
    
김철수.광물확인()
    김철수씨는 현재 금 1개, 은 1개, 다이아몬드 1개를 가지고 있습니다.

```

`````python
# write your code here!!

# sleep(n) 을 하면 n초간 코드 진행을 지연할 수 있습니다.

import time
# from time import sleep

class 광부():
    
    def __init__(self, 이름, 목표):
        self.현재 = [0, 0, 0]
        self.이름 = 이름
        self.목표 = 목표
        
    def 채굴(self):
        self.material = input("무슨 광물을 채굴할까?")
        if self.material == "금":
            self.현재[0] += 1
        elif self.material == "은":
            self.현재[1] += 1
        else:
            self.현재[2] += 1
        
        print("뚝딱뚝뚝....(3초)")
#         time.sleep(3)
        sleep(3)
        print(f"{self.material}을(를) 채굴했습니다.")
        
    def 광물확인(self):
        print(f"{self.이름}씨는 현재 금 {self.현재[0]}개, 은 {self.현재[1]}개, 다이아몬드 {self.현재[2]}개를 가지고 있습니다.")
        if self.현재[0] == self.목표[0] and self.현재[1] == self.목표[1] and self.현재[2] == self.목표[2]: 
            print("축하합니다. 목표를 달성했습니다!")
        else:
            print(f"목표까지 금 {1 - self.현재[0]}개, 은 {1 - self.현재[1]}개, 다이아몬드 {1 - self.현재[2]}개가 필요합니다.")
`````

`````python
김철수 = 광부(이름='김철수', 목표=[1,1,1])

김철수.채굴()
김철수.광물확인()
김철수.채굴()
김철수.채굴()
김철수.광물확인()
`````

`````python
# Session

# return self.__init__(self.money)

class ATM:
    
    def __init__(self, money):
        self.money = money
        
        num = int(input("""
        1. 입금
        2. 출금
        3. 잔액확인
        4. 거래종료
        선택(1~4):"""))
        
        if num == 1:
            self.save()
        elif num == 2:
            self.withdraw()
        elif num == 3:
            self.check_balance()
        elif num == 4:
            self.exit()
    
    def save(self):
        self.money += int(input("입금액 :"))
        print(f"입금 후 잔액: {self.money}원")
        
        return self.__init__(self.money)
    
    def withdraw(self):
        self.money -= int(input("출금액 :"))
        print(f"출금 후 잔액: {self.money}원")
        
        return self.__init__(self.money)

    def check_balance(self):
        print(f"잔액: {self.money}원")
        
        return self.__init__(self.money)
        
    def exit(self):
        
        return # terminate
`````

```python
잔액이 부족하여 송금이 불가능합니다.

```

`````python
# write your code here!!

class Bank:
    account = {}
    
    def __init__(self, name, balance):
        Bank.account[name] = balance

    def transfer(self, target, money):
        self.target = target
        self.sender = "철수"
            
        self.money = money
        if Bank.account[self.sender] >= money:
            Bank.account[target] += money
            Bank.account[self.sender] -= money
            print(f"{self.sender}가 {self.target}에게 {self.money}원을 송금했습니다. {self.sender}의 잔액은 {Bank.account['철수']}원입니다.")
            
        else:
            print("잔액이 부족하여 송금이 불가능합니다.")
        
    def check_balance(self):
        print(f"짱구의 잔액은 {Bank.account['짱구']}원입니다.")
        

철수 = Bank(name = '철수', balance = 10000)
짱구 = Bank(name = '짱구', balance = 20000)


철수.transfer(target = '짱구', money = 2000)
짱구.check_balance()

철수.transfer(target = '짱구', money = 8000)
`````

`````python
# Session

class Bank:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        
    def transfer(self, target, money):
        
        if self.balance >= money:
            self.balance -= money
            target.balance += money
            return print(f"{self.name}가 {target.name}에게 {money}원을 송금했습니다. {self.name}의 잔액은 {self.balance}원입니다.")
        
        else:
            return print("잔액이 부족하여 송금이 불가능합니다.")
        
    def check_balance(self):
        return print(f'{self.name}의 잔액은 {self.balance}원입니다.')
`````

`````python
철수 = Bank('철수', 10000)
짱구 = Bank('짱구', 20000)

철수.transfer(target = 짱구, money = 2000)
짱구.check_balance()

철수.transfer(target = 짱구, money = 50000)
`````

`````python
# write your code here!!

class Human():
    
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    
    def print_info(self):
        print(f"""
        이름 : {self.name}
        번호 : {self.phone}
        """)

class Student(Human):
    
    def __init__(self, name, phone, college, student_id):
        super().__init__(name, phone)
        self.college = college
        self.student_id = student_id
        
    def print_info(self):
        print(f"""
        이름 : {self.name}
        번호 : {self.phone}
        학교 : {self.college}
        학번 : {self.student_id}
        """)

철수 = Student(name = "철수", phone= "010-1234-1234", college = "Harvard", student_id = 1234)
철수.print_info()
`````
